```python
import pandas as pd
import numpy as np
import os 
```

```python
# 导入文件路径
abs_ic_res_folder_path = r'D:\QUANT_GAME\python_game\factor\factor_lab_demo\random_factor_ic_searching\abs_ic'
yearly_abs_ic_res_folder_path = r'D:\QUANT_GAME\python_game\factor\factor_lab_demo\random_factor_ic_searching\yearly_abs_ic'
# 获取文件列表
abs_ic_file_name_list = os.listdir(abs_ic_res_folder_path)
yearly_abs_ic_file_name_list = os.listdir(yearly_abs_ic_res_folder_path)
# 获取文件路径列表
abs_ic_file_path_list = [os.path.join(abs_ic_res_folder_path, file_name) for file_name in abs_ic_file_name_list]
yearly_abs_ic_file_path_list = [os.path.join(yearly_abs_ic_res_folder_path, file_name) for file_name in yearly_abs_ic_file_name_list]
```

```python
def read_abs_ic_file(file_path_list):
    abs_ic_res_list = []
    for file_path in file_path_list:
        abs_ic_res = pd.read_pickle(file_path)
        print('abs_ic_res len:', abs_ic_res.__len__())
        abs_ic_res_list.extend(abs_ic_res)
        print('abs_ic_res_list len:', abs_ic_res_list.__len__())
    return abs_ic_res_list
```

```python
def read_yearly_abs_ic_file(file_paths_list):
    yearly_abs_ic_res_dict = dict()
    for file_path in file_paths_list:
        yearly_abs_ic_dict = pd.read_pickle(file_path)
        yearly_abs_ic_df = pd.DataFrame.from_dict(yearly_abs_ic_dict, orient='index')
        print('df_shape', yearly_abs_ic_df.shape)
        for year in yearly_abs_ic_df.index:
            if year not in yearly_abs_ic_res_dict.keys():
                yearly_abs_ic_res_dict[year] = []
            yearly_abs_ic_res_dict[year].extend(list(yearly_abs_ic_df.loc[year].values))
        yearly_abs_ic_res_dict_df = pd.DataFrame.from_dict(yearly_abs_ic_res_dict, orient='index')
        print('yearly_abs_ic_res_dict_df_shape:', yearly_abs_ic_res_dict_df.shape)
        del yearly_abs_ic_res_dict_df
    return yearly_abs_ic_res_dict
```

```python
yearly_abs_ic_res = read_yearly_abs_ic_file(yearly_abs_ic_file_path_list)
```

```python
yearly_abs_ic_res_df = pd.DataFrame.from_dict(yearly_abs_ic_res, orient='index').T
```
