```python
import pandas as pd
import datetime 
```

```python
date_bar = pd.read_pickle(r'D:\QUANT_GAME\python_game\factor\factor_lab_demo\get_date_bar\date_bar_post.pkl')
date_bar_tmp = date_bar.copy().reset_index()
```

```python
stks_num_every_date = date_bar_tmp.groupby('date',group_keys=False)['idxs'].count()
```

```python
date_idx = stks_num_every_date.index
datetime_date_idx = [datetime.datetime.strptime(str(date),'%Y%m%d') for date in date_idx]
```

```python
stks_num_every_date.index = datetime_date_idx
# 绘制图片，y轴为股票数量
stks_num_every_date.plot()
```
